```python
import pandas as pd

df = pd.read_csv("MOCK_DATA.csv")

df
```

```python
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.xlabel("Year")
plt.ylabel("Income")

plt.scatter(df.year, df.income, color='red', marker='*')
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['income'])
plt.plot(df.year, reg.predict(df[['year']]), color='green')
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
prediction = reg.predict(pd.DataFrame([[2020]], columns=['year']))

print("Predicted Income in 2020:", prediction)
```
